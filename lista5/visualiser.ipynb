```python
import matplotlib.pyplot as plt
```

LCS

```python
x = []
y = []

with open('data/lcs.txt') as file:
    for line in file.readlines():
        line = line.strip().split()
        x.append(float(line[0]))
        y.append(float(line[1]))

plt.scatter(x, y)
plt.xlabel('Length of sequence')
plt.ylabel('Number of operations')
plt.title('Longest Common Subsequence')
plt.show()
```

HEAPS

```python
SERIES_NUMBER = 5
N_VALUES = [500, 1000]
ALGS = ['binomial', 'fibonacci']
colors = ['magenta', 'yellow', 'red', 'green', 'blue']
```

```python
MAX_Y_VAL = 40

def proccess_alg(alg, n):
    series = []

    for k in range(SERIES_NUMBER):
        with open(f'data/{alg}_heap_{n}-{k}.txt') as file:
            x = []
            y = []
            for line in file.readlines():
                line = line.strip().split()
                x.append(int(line[0]))
                if float(line[1]) < MAX_Y_VAL:
                    y.append(float(line[1]))
                else:
                    y.append(MAX_Y_VAL)
                    plt.text(int(line[0]) + 50, MAX_Y_VAL - 0.5, f'y = {float(line[1])}', fontsize=10)

            series.append([x, y])

    for i, s in enumerate(series):
        plt.scatter(s[0], s[1], c=colors[i], marker='x', s=20)

    plt.xlabel('Instruction number; first half is insertion, second half is decrease key')
    plt.ylabel('Number of comparisions')
    plt.title(f'Comparisions Number Per Instruction For {alg.capitalize()} Heap; n={n}')
    plt.show()
```

```python
for alg in ALGS:
    for n in N_VALUES:
        proccess_alg(alg, n)
```

COMPARE HEAPS

```python
bin_x = []
bin_y = []
fib_x = []
fib_y = []

with open('data/binomial_heap.txt') as file:
    for line in file.readlines():
        line = line.strip().split()
        bin_x.append(int(line[0]))
        bin_y.append(float(line[1]))

with open('data/fibonacci_heap.txt') as file:
    for line in file.readlines():
        line = line.strip().split()
        fib_x.append(int(line[0]))
        fib_y.append(float(line[1]))

plt.scatter(bin_x, bin_y, s=20, c=colors[3], label='Binomial Heap')
plt.scatter(fib_x, fib_y, s=20, c=colors[4], label='Fibonacci Heap')
plt.title('Total Comparision Number: Binomial and Fibonacci Heap')
plt.xlabel('Number of elements in one of 2 heaps before union')
plt.ylabel('Number of comparisions')
plt.legend()
plt.show()
```
